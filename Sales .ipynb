```python
import pandas as pd
```

```python
customer = pd.read_csv('olist_customers_dataset.csv')
order_item = pd.read_csv('olist_order_items_dataset.csv')
order_review = pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
product = pd.read_csv('olist_products_dataset.csv')
seller = pd.read_csv('olist_sellers_dataset.csv')
```

```python
#data info
```

```python
all_datasets_name = [customer, order_item, order_review, orders, product, seller]
title_show = ['customer', 'order_item', 'order_review', 'orders', 'product', 'seller']
```

```python
data_info = pd.DataFrame({},)
data_info['dataset_name'] = title_show

data_info['info_columns'] = [len(dataframe.columns) for dataframe in all_datasets_name]
data_info['columns_name'] = [', '.join(list(dataframe.columns)) for dataframe in all_datasets_name]
data_info['total_rows'] = [len(dataframe) for dataframe in all_datasets_name]

data_info.style.background_gradient()
```

```python
#checking na value
```

```python
info_df_na = pd.DataFrame({},)

info_df_na['dataset'] = title_show

#column of name columns in dataset
info_df_na['all_cols_name'] = [', '.join([col for col, null in dataframe.isnull().sum().items()]) for dataframe in all_datasets_name]

#total number of columns in dataset
info_df_na['total_cols_count'] = [dataframe.shape[1] for dataframe in all_datasets_name]

#counting total na values
info_df_na['total_null_count'] = [dataframe.isnull().sum().sum() for dataframe in all_datasets_name]

#total number of columns in dataset with null values
info_df_na['na_cols_count'] = [len([col for col, null in dataframe.isnull().sum().items() if null > 0]) for dataframe in all_datasets_name]

#column of name of columns in dataset with null values
info_df_na['na_cols_name'] = [', '.join([col for col, null in dataframe.isnull().sum().items() if null > 0]) for dataframe in all_datasets_name]

info_df_na.style.background_gradient()
```
